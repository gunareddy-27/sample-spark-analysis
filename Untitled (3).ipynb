```python
# Cell 1: imports and Spark session
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, isnan, count, lit
from pyspark.sql.types import StringType, DoubleType
from pyspark.ml import Pipeline
from pyspark.ml.feature import (
    StringIndexer, OneHotEncoder, VectorAssembler, StandardScaler, PCA
)
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier, GBTClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml.clustering import KMeans
from pyspark.ml.stat import Correlation
import pyspark.sql.functions as F

spark = SparkSession.builder \
    .appName("AdvancedAnalysisAndML") \
    .config("spark.sql.execution.arrow.enabled", "true") \
    .getOrCreate()

DATA_PATH = "crocodile_dataset.csv"
```

```python
# Cell 2: load data and automatic label detection
df = spark.read.option("header", True).option("inferSchema", True).csv(DATA_PATH)
print("Rows:", df.count())
df.printSchema()
display(df.limit(5).toPandas())  # in notebook; otherwise use df.show(5)

# Auto-detect label column
candidate_labels = [c for c in df.columns if c.lower() in ("label","target","class","y")]
if candidate_labels:
    label_col = candidate_labels[0]
else:
    # fallback to last column
    label_col = df.columns[-1]

print("Detected label column:", label_col)
```

```python
# Cell 3: Basic EDA - missing values, types, summary
# Missing values count by column
missing = df.select([count(when(col(c).isNull() | isnan(col(c)) | (col(c) == ''), c)).alias(c) for c in df.columns])
display(missing.toPandas())

# Data types and numeric columns
dtypes = dict(df.dtypes)
numeric_cols = [c for c,t in dtypes.items() if t in ('int','double','float','bigint','long','decimal')]
cat_cols = [c for c in df.columns if c not in numeric_cols and c != label_col]
# ensure label not in features
if label_col in numeric_cols:
    numeric_cols.remove(label_col)
if label_col in cat_cols:
    cat_cols.remove(label_col)

print("Numeric columns:", numeric_cols)
print("Categorical/string columns:", cat_cols)

# Summary stats for numeric
display(df.select(numeric_cols).describe().toPandas())
```

```python
# Cell 4: Correlation matrix for numeric features (if enough numeric cols)
if len(numeric_cols) >= 2:
    assembler_corr = VectorAssembler(inputCols=numeric_cols, outputCol="corr_features")
    corr_df = assembler_corr.transform(df.select(numeric_cols).na.drop())
    corr_mat = Correlation.corr(corr_df, "corr_features").head()[0]
    print("Correlation matrix (row-major):")
    print(corr_mat)  # DenseMatrix
else:
    print("Not enough numeric columns for correlation.")
```

```python
# Cell 5: Missing-value handling and cleanup
# Strategy:
# - For numeric: impute with median
# - For categorical: fill with "MISSING"
from pyspark.ml.feature import Imputer

# Fill categorical nulls
for c in cat_cols:
    df = df.withColumn(c, when(col(c).isNull() | (col(c) == ''), lit("MISSING")).otherwise(col(c)))

# Numeric imputer
if numeric_cols:
    imputer = Imputer(inputCols=numeric_cols, outputCols=[f"{c}_imp" for c in numeric_cols]).setStrategy("median")
    df = imputer.fit(df).transform(df)
    # replace numeric_cols with their _imp versions for modeling
    numeric_imputed = [f"{c}_imp" for c in numeric_cols]
else:
    numeric_imputed = []

# If label is string, index it; else convert to double
label_type = dict(df.dtypes)[label_col]
if label_type not in ('int','double','float','bigint'):
    # We'll index label
    label_indexer = StringIndexer(inputCol=label_col, outputCol="label", handleInvalid="keep")
    df = label_indexer.fit(df).transform(df)
else:
    df = df.withColumn("label", col(label_col).cast("double"))
```

```python
# Cell 6: Categorical encoding (StringIndexer + OneHotEncoder) and assembling features
indexers = [StringIndexer(inputCol=c, outputCol=f"{c}_idx", handleInvalid="keep") for c in cat_cols]
ohe = OneHotEncoder(inputCols=[f"{c}_idx" for c in cat_cols],
                    outputCols=[f"{c}_ohe" for c in cat_cols],
                    handleInvalid="keep")

feature_inputs = []
feature_inputs += numeric_imputed
feature_inputs += [f"{c}_ohe" for c in cat_cols]

assembler = VectorAssembler(inputCols=feature_inputs, outputCol="features_raw", handleInvalid="keep")
scaler = StandardScaler(inputCol="features_raw", outputCol="features", withStd=True, withMean=False)

stages = []
stages += indexers
stages += [ohe, assembler, scaler]

pipeline_feat = Pipeline(stages=stages)
pipeline_model = pipeline_feat.fit(df)
df_transformed = pipeline_model.transform(df)

print("Transformed columns:", df_transformed.columns)
display(df_transformed.select("label", "features").limit(5).toPandas())
```

```python
# Cell 1: Spark Session
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, count, lit, isnan
from pyspark.ml.feature import Imputer

spark = SparkSession.builder.appName("DataCleaningAndViz").getOrCreate()

DATA_PATH = "crocodile_dataset.csv"

# Load data
df = spark.read.option("header", True).option("inferSchema", True).csv(DATA_PATH)
print("Rows:", df.count())
df.printSchema()
```

```python
# Cell 2: Check missing values
missing = df.select([
    count(when(col(c).isNull() | (col(c) == "") | isnan(col(c)), c)).alias(c)
    for c in df.columns
])
missing.show()
```

```python
# Cell 3: Handle missing values
# Separate numeric and categorical columns
numeric_cols = [c for c, t in df.dtypes if t in ("int", "double", "float", "bigint", "long")]
cat_cols = [c for c in df.columns if c not in numeric_cols]

# Fill categorical missing with "MISSING"
for c in cat_cols:
    df = df.withColumn(c, when(col(c).isNull() | (col(c) == ""), lit("MISSING")).otherwise(col(c)))

# Impute numeric missing with median
if numeric_cols:
    imputer = Imputer(inputCols=numeric_cols, outputCols=[f"{c}_imp" for c in numeric_cols]).setStrategy("median")
    df = imputer.fit(df).transform(df)
    # Replace numeric cols with imputed versions
    for c in numeric_cols:
        df = df.drop(c).withColumnRenamed(f"{c}_imp", c)

df.show(5)
```

```python
# Cell 4: Remove duplicates
df = df.dropDuplicates()
print("Rows after dropping duplicates:", df.count())
```

```python
# Cell 5: Outlier treatment (optional - Z-score method for numeric columns)
from pyspark.sql.functions import mean, stddev

for c in numeric_cols:
    stats = df.select(mean(col(c)), stddev(col(c))).first()
    mean_val, std_val = stats[0], stats[1]
    if std_val:  # avoid div by zero
        df = df.withColumn(c,
            when(((col(c) - mean_val) / std_val) > 3, mean_val)  # cap outliers
            .when(((col(c) - mean_val) / std_val) < -3, mean_val)
            .otherwise(col(c))
        )
```

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

spark = SparkSession.builder \
    .appName("AdvancedDataCleaning_Fixed") \
    .config("spark.sql.execution.arrow.enabled", "true") \
    .getOrCreate()

DATA_PATH = "crocodile_dataset.csv"  # adjust if needed

df = spark.read.option("header", True).option("inferSchema", True).csv(DATA_PATH)
print("Initial rows:", df.count())
df.printSchema()
```

```python
# standardize column names
df = df.toDF(*[c.strip().lower().replace(" ", "_") for c in df.columns])

# detect numeric / categorical
numeric_cols = [c for c, t in df.dtypes if t in ("int", "double", "float", "bigint", "long")]
cat_cols = [c for c in df.columns if c not in numeric_cols]

print("Numeric cols:", numeric_cols)
print("Categorical cols:", cat_cols)

# cache for repeated use
df = df.cache()
```

```python
from pyspark.ml.feature import Imputer

# Fill categorical missing with "missing"
for c in cat_cols:
    df = df.withColumn(c, F.when(F.col(c).isNull() | (F.col(c) == ""), F.lit("missing")).otherwise(F.col(c)))

# Impute numeric columns with median (creates _imp columns then replace)
if numeric_cols:
    imputer = Imputer(inputCols=numeric_cols,
                      outputCols=[f"{c}_imp" for c in numeric_cols]).setStrategy("median")
    df = imputer.fit(df).transform(df)
    # drop originals and rename imputed
    for c in numeric_cols:
        df = df.drop(c).withColumnRenamed(f"{c}_imp", c)

df = df.cache()
print("After missing value handling, rows:", df.count())
```

```python
# Option A: IQR-based capping per numeric column (safer than dropping)
for c in numeric_cols:
    # compute quantiles; keep relative error small
    q1, q3 = df.approxQuantile(c, [0.25, 0.75], 0.01)
    iqr = q3 - q1
    if iqr is None:
        continue
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    # cap values outside bounds to the nearest bound
    df = df.withColumn(c, F.when(F.col(c) < lower, F.lit(lower))
                             .when(F.col(c) > upper, F.lit(upper))
                             .otherwise(F.col(c)))

# Optional: Z-score capping (commented out; use one method at a time)
# for c in numeric_cols:
#     stats = df.select(F.mean(c).alias("m"), F.stddev(c).alias("s")).first()
#     m, s = stats["m"], stats["s"]
#     if s and s != 0:
#         df = df.withColumn(c,
#             F.when(((F.col(c) - m) / s) > 3, F.lit(m))
#              .when(((F.col(c) - m) / s) < -3, F.lit(m))
#              .otherwise(F.col(c))
#         )

df = df.cache()
print("Outlier treatment done.")
```

```python
from pyspark.sql.functions import trim, lower, regexp_replace

for c in cat_cols:
    df = df.withColumn(c, trim(lower(F.col(c))))
    # remove characters that are not letter, number, or underscore
    df = df.withColumn(c, regexp_replace(F.col(c), "[^a-z0-9_]", ""))

df = df.cache()
print("String cleaning completed.")
```

```python
print("Final schema:")
df.printSchema()
print("Sample rows:")
df.show(10, truncate=False)

OUTPUT_PATH = "/mnt/data/crocodile_dataset_cleaned.csv"
# write as single CSV file if dataset is small; otherwise write partitioned folder
# If you want a single file, you can coalesce(1) but beware memory for large df.
df.coalesce(1).write.mode("overwrite").option("header", True).csv(OUTPUT_PATH)
print(f"Saved cleaned data to {OUTPUT_PATH}")
```

```python
from pyspark.sql import functions as F

# Row-wise sum and avg for numeric columns
if numeric_cols:
    df = df.withColumn("numeric_sum", sum([F.col(c) for c in numeric_cols]))
    df = df.withColumn("numeric_avg", (F.col("numeric_sum") / len(numeric_cols)))

# Flag if any categorical value was originally "missing"
if cat_cols:
    df = df.withColumn("has_missing_cat", F.when(
        sum([F.when(F.col(c) == "missing", 1).otherwise(0) for c in cat_cols]) > 0,
        1
    ).otherwise(0))

# Length of concatenated categorical values
if cat_cols:
    df = df.withColumn("row_length", F.length(F.concat_ws("_", *[F.col(c) for c in cat_cols])))

df = df.cache()
print("Added engineered columns.")
df.show(5, truncate=False)
```

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler

# Index categorical cols
indexers = [StringIndexer(inputCol=c, outputCol=f"{c}_idx", handleInvalid="keep") for c in cat_cols]

# One-hot encode
encoders = [OneHotEncoder(inputCol=f"{c}_idx", outputCol=f"{c}_ohe") for c in cat_cols]

# Collect feature columns (numeric + ohe + engineered)
feature_cols = numeric_cols + ["numeric_sum", "numeric_avg", "has_missing_cat", "row_length"] + [f"{c}_ohe" for c in cat_cols]

assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")

print("Feature engineering pipeline ready.")
```

```python
from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline

# Define target (create one if not present)
target_column = "label"
if target_column not in df.columns:
    df = df.withColumn(target_column, F.when(F.col("numeric_avg") > df.agg(F.avg("numeric_avg")).first()[0], 1).otherwise(0))

# Index label
label_indexer = StringIndexer(inputCol=target_column, outputCol="label_idx")

# Logistic Regression model
lr = LogisticRegression(featuresCol="features", labelCol="label_idx", maxIter=20)

# Build pipeline
pipeline = Pipeline(stages=indexers + encoders + [assembler, label_indexer, lr])

# Train/test split
train_df, test_df = df.randomSplit([0.7, 0.3], seed=42)
model = pipeline.fit(train_df)

# Predictions
predictions = model.transform(test_df)
predictions.select("features", "label_idx", "prediction", "probability").show(10, truncate=False)
```

```python
from pyspark.ml.regression import RandomForestRegressor

# Define regression target
reg_target = "numeric_sum"

# Random Forest Regressor
rf = RandomForestRegressor(featuresCol="features", labelCol=reg_target, numTrees=50)

# Pipeline without label indexer (since regression)
reg_pipeline = Pipeline(stages=indexers + encoders + [assembler, rf])

train_df, test_df = df.randomSplit([0.7, 0.3], seed=123)
reg_model = reg_pipeline.fit(train_df)

# Predictions
reg_preds = reg_model.transform(test_df)
reg_preds.select("features", reg_target, "prediction").show(10, truncate=False)
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator, RegressionEvaluator

# Classification evaluation
clf_eval = BinaryClassificationEvaluator(labelCol="label_idx", rawPredictionCol="rawPrediction")
auc = clf_eval.evaluate(predictions)
print("Classification AUC:", auc)

# Regression evaluation
reg_eval = RegressionEvaluator(labelCol=reg_target, predictionCol="prediction", metricName="rmse")
rmse = reg_eval.evaluate(reg_preds)
print("Regression RMSE:", rmse)
```

```python
from pyspark.sql.functions import col

# Load dataset
df = spark.read.csv("crocodile_dataset.csv", header=True, inferSchema=True)

# Rename columns to simpler names
df = df.withColumnRenamed("Observed Length (m)", "Length") \
       .withColumnRenamed("Observed Weight (kg)", "Weight")

# Add BMI-like ratio column
df = df.withColumn("BMI_Like", col("Weight") / (col("Length") * col("Length")))

# Show result
df.select("Weight", "Length", "BMI_Like").show(10, truncate=False)
```

```python
from pyspark.sql.functions import when, col

# Add AgeGroup column based on Length
df = df.withColumn(
    "AgeGroup",
    when(col("Length") < 2.0, "Juvenile")
    .when((col("Length") >= 2.0) & (col("Length") < 3.5), "Sub-Adult")
    .when(col("Length") >= 3.5, "Adult")
    .otherwise("Unknown")   # fallback for missing values
)

# Show some samples
df.select("Length", "AgeGroup").show(10, truncate=False)
```
